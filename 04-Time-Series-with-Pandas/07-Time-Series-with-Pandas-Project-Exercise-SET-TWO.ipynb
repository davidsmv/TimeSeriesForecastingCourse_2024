**Create a bar plot showing the average value in millions of dollars per year**

**What year had the biggest increase in mean value from the previous year's mean value? (Lots of ways to get this answer!)**

[HINT for a useful method](https://pandas.pydata.org/pandas-docs/version/0.21/generated/pandas.DataFrame.idxmax.html)

```python
# CODE HERE
(df.resample(rule='YE').mean() - df.resample(rule='YE').mean().shift(1)).idxmax()
```

**Plot out the yearly rolling mean on top of the original data. Recall that this is monthly data and there are 12 months in a year!**

```python
# CODE HERE
# Create a figure and axis object
fig, ax = plt.subplots(figsize=(12,5))

# Plot the original data with a custom label
ax.plot(df.index, df, label='Original Data')

# Plot the rolling mean on the same axis with a custom label
ax.plot(df.index, df.rolling(window=12).mean(), label='Rolling Mean (12 months)')

# Adjust the x-axis to be tight
ax.autoscale(axis='x', tight=True)

# Show the legend with the new labels
ax.legend()

# Show the plot
plt.show()
```

**BONUS QUESTION (HARD).**

**Some month in 2008 the value peaked for that year. How many months did it take to surpass that 2008 peak? (Since it crashed immediately after this peak) There are many ways to get this answer. NOTE: I get 70 months as my answer, you may get 69 or 68, depending on whether or not you count the start and end months. Refer to the video solutions for full explanation on this.**

```python
#CODE HERE
df_month = df.resample(rule='MS').sum()
max_2008 = df_month["2008-01-01":"2008-12-31"].max()

(df_month["2008-01-01":"2008-12-31"].idxmax() - (df_month[(df_month.index > (df_month["2008-01-01":"2008-12-31"].idxmax()[0])) & (df_month["UMTMVS"] > float(max_2008))].index[0]))
```

```python
import pandas as pd

# Resample the data by month and sum the values
df_month = df.resample(rule='MS').sum()

# Get the maximum value in 2008 for UMTMVS column
max_2008 = df_month["2008-01-01":"2008-12-31"]["UMTMVS"].max()

# Get the index of the maximum value in 2008 for UMTMVS column
max_index_2008 = df_month["2008-01-01":"2008-12-31"]["UMTMVS"].idxmax()

# Find the next index where the condition is met (index is greater than max_index_2008 and value is greater than max_2008)
next_index = df_month[(df_month.index > max_index_2008) & (df_month["UMTMVS"] > float(max_2008))].index[0]

# Calculate the difference in months
difference_in_months = (next_index.year - max_index_2008.year) * 12 + (next_index.month - max_index_2008.month)

difference_in_months
```
